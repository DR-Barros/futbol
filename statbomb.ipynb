```python
events = sb.events(match_id=3939981, split=True)
```

```python
eventos = []
for event in events:
    eventos.append(event)
eventos.pop(0)
print(eventos)
```

```python
events["shots"]
```

```python
import pandas as pd
location = []
for event in events:
    try:
        location.append(events[event]["location"])
    except:
        pass

df = pd.concat(location)
df = pd.DataFrame(df.tolist(), columns=['x', 'y'])
print(df.shape,"\n" ,df.head())
print(f"x: {max(df.x)}, {min(df.x)}")
print(f"y: {max(df.y)}, {min(df.y)}")
```

```python
data = []
for event in eventos:
    try:
        evento = events[event]
        if not ("location" in evento.columns):
            evento["location"] = None
        data.append(evento[["period", "timestamp", "type", "location", "possession_team","team"]])
    except Exception as e:
        print(f"{event} fallo por {e}")

data = pd.concat(data)
#ordenamos los eventos por el tiempo
data = data.sort_values(by=["period", "timestamp"])
data = data.reset_index(drop=True)
```

```python
data
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time

# Activamos modo interactivo
plt.ion()

fig, ax = plt.subplots()
sc = ax.scatter([], [])
ax.set_xlim(0, 120)
ax.set_ylim(0, 80)

def update(row):
    if row["location"] is None:
        ax.set_title(f"Periodo: {row['period']} | Tiempo: {row['timestamp']}s")
        plt.draw()
        plt.pause(0.1)
        return
    
    x = row["location"][0] 
    y = row["location"][1] 
    positions = np.array([[x, y]])

    if not np.isnan(x) and not np.isnan(y):
        sc.set_offsets(positions)

    ax.set_title(f"Periodo: {row['period']} | Tiempo: {row['timestamp']}s")
    plt.draw()
    plt.pause(0.1)  # pausa breve para ver el cambio

# Iteramos sobre las filas y actualizamos el gráfico
for _, row in data.iterrows():
    update(row)

plt.ioff()
plt.show()

```
